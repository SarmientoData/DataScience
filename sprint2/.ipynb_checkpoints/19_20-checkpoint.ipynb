### Importacion de librerias y carga de dataset

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("/content/DSONLINE/BLOQUE2/19-20/data.csv")
```

```
df.head()
```

### Pairplot para detectar correlaciones entre variables

```
sns.pairplot(df)
```

### Conteo de valores faltantes por columna

```
for col in df.columns:
  r = df[df[col]== "?"][col].count()
  print(col, r)
```

### Cut de una variable NUMERICA para analizar valores faltantes por rango

```
sns.distplot(df["anual-salary"])
```

```
df["salary_cut"]= pd.cut(df["anual-salary"], bins=10 , labels=False)
```

```
df.head()
```

```
salary_cut = df[df.workclass == "?"]["salary_cut"].value_counts().sort_index()
salary_cut
```

```
sns.lineplot(salary_cut.index, salary_cut)
```

```
###Completado de valores faltantes por 0, esto se hace para no tener estos valores en el proximo punto de encoding
for col in df.columns:
  df.loc[df[col] =="?", col] = 0

df.fillna(0, inplace=True)
```
